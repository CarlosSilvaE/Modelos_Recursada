```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [10, 20, 30],
    'C': [100, 200, 300]
})

print(df)
```

```python
#USANDO LAMBDA CON AXIS = 0 (POR COLUMNAS)
df.apply(lambda col: col.max(), axis = 0)
```

```python
#USANDO LAMBDA CON AXIS = 1 (POR FILAS)
df.apply(lambda row: row.sum(), axis = 1)
```
