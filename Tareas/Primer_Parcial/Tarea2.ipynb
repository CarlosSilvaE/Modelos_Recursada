```python
import matplotlib.pyplot as plt
import random
import pandas as pd
import math as ma
import numpy as ny
```

```python
#1. GENERAR 20 NUMEROS RANDOMS CON SU RAIZ CUADRADA

numeros = []
raices = []

for _ in range(20):
    n = random.randint(0, 100)
    numeros.append(n)
    raices.append(ma.isqrt(n)) 
```

```python
plt.scatter(numeros, raices)
plt.xlabel('Numeros')
plt.ylabel('Raices cuadradas')
plt.title('Raices de numeros aleatorios del 0 al 100')

plt.show()
```

```python
#2. SACAR PRODUCTO ESCALAR DE DOS MATRICES 3X3

ancho1 = 3 ; largo1 = 3
ancho2 = 3 ; largo2 = 3

matriz1 = [[random.randint(-1000, 1000) for j in range(largo1)] for i in range(ancho1)]
matriz2 = [[random.randint(-1000, 1000) for a in range(largo2)] for b in range(ancho2)]

print("Matrices con datos aleatorios\n")
for f in matriz1:
    print(f)

print("")

for g in matriz2:
    print(g)

print("\nProducto escalar\n")
resultado = ny.dot(matriz1, matriz2)
print(resultado)


```
