```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#CREAR ARRAYS Y DATAFRAME
dias = ["Lun", "Mar", "Mie", "Jue", "Vie", "Sab", "Dom"]

ventas = [120, 80, 150, 200, 90, 170, 130]

df = pd.DataFrame({'dias': dias, 'ventas': ventas})

df
```

```python
#SACAR PROMEDIO DE VENTAS
promedio = sum(ventas) / len(ventas)

#CREAR GRAFICA DE BARRAS USANDO VALORES DE DIAS COMO X & VENTAS COMO Y
cnt = 0

barras = plt.bar(df['dias'], df['ventas'], color = 'darkgreen')
plt.xlabel('Dias')
plt.ylabel('Ventas')
plt.title('Ventas Semanales de la Tienda')
plt.bar_label(barras, fmt='{:,.0f}')

#SACAR LINEA DE PROMEDIO DE VENTAS
plt.axhline(y=promedio, linestyle='--', color = "red")

for i in df['ventas']:
    plt.Text(cnt, i + 0.1, ha = "center")
    cnt+=1
plt.show()
```
