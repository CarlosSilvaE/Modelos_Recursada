Realizar la multiplicación de matrices que para alguna posible solucion de  cadena de markov, en Python, la idea es iterar y resolver la matriz estable y parar , al final mostrar esos resultados estables a 4 decimales. La entrada de la matriz puede ser fija o variable.  Ejemplo a resolver, subirlo(cargar) en un jupyter notebook. 

```python
import numpy as np
m1 = np.matrix([[.25, .50, .25],[.05, .90, .05], [.10, .50, .40]])
m1
```

```python
m_aux = np.matrix([[], [], []])
m_aux_anterior = np.matrix([[], [], []])
```

```python
#DETECTAR MATRIZ ESTABLE
for i in range(1000):
    if i == 0:
        m_aux = np.dot(m1, m1)
        print(m_aux)
        print("1-------------", i + 1)
    else:
        y = i + 1
        m_aux_anterior = m_aux
        m_aux = np.dot(m_aux, m1)
        print(m_aux)
        if (np.array_equal(m_aux, m_aux_anterior)):
            break
        print(y, "-------", i + 1)
print()
print("La matriz estable es:\n", np.round(m_aux, 4))
```
