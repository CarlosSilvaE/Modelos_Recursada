# Cadenas de Markov

1.- (Cadena de MARKOV) El departamento de estudios de mercado de una fábrica estima que el 20% de la gente que compra un producto un mes, no lo comprará el mes siguiente.

Además, el 30% de quienes no lo compren un mes lo adquirirá al mes siguiente.

En una población de 1000 individuos, 100 compraron el producto el primer mes. ¿Cuántos lo comprarán al mes próximo? ¿Y dentro de dos meses?

```python
""" 
    NC  SC
NC  0.7 0.3
SC  0.2 0.8
"""
import numpy as np
m1 = np.matrix([[0.7, 0.3], [0.2, 0.8]])
m1
```

```python
m_aux = np.matrix([[], []])
m_aux_anterior = np.matrix([[], []])
```

```python
#DETECTAR MATRIZ ESTABLE
for i in range(1000):
    if i == 0:
        m_aux = np.dot(m1, m1)
        print(m_aux)
        print("1-------------", i + 1)
    else:
        y = i + 1
        m_aux_anterior = m_aux
        m_aux = np.dot(m_aux, m1)
        print(m_aux)
        if (np.array_equal(m_aux, m_aux_anterior)):
            break
        print(y, "-------", i + 1)
print()
print("La matriz estable es:\n", np.round(m_aux, 4))
```

```python

poblacion_inicial = np.matrix([900, 100])
poblacion_inicial
```

```python
#¿Cuantos lo comprararn el mes proximo?
poblacion_proxima = np.dot(poblacion_inicial, m1)
print("La poblacion del proximo mes es:\n", np.round(poblacion_proxima, 4))
```

```python
#¿Cuantos lo comprararn en 2 meses?
poblacion_dos = np.dot(poblacion_proxima, m1)
print("La poblacion del proximo mes es:\n", np.round(poblacion_dos, 4))
```

Los datos se

```python
#DETECTAR MATRIZ ESTABLE
for i in range(1000):
    if i == 0:
        m_aux = np.dot(m1, m1)
        print(m_aux)
        print("1-------------", i + 1)
    else:
        y = i + 1
        m_aux_anterior = m_aux
        m_aux = np.dot(m_aux, m1)
        print(m_aux)
        if (np.array_equal(m_aux, m_aux_anterior)):
            break
        print(y, "-------", i + 1)
print()
print("La matriz estable es:\n", np.round(m_aux, 4))
print("-------------------")
print(np.round(m_aux, 3))
```

3.- Problema Planteado - Teoria de colas

Una empresa tiene un centro de servicio donde los clientes llegan en promedio 8 clientes por hora (λ\lambda) siguiendo un proceso de distribución de Poisson. Para atender a los clientes, hay tres servidores que trabajan de manera paralela y con un tiempo de servicio de 11 clientes por hora(μ\muμ.) que sigue una distribución exponencial.

Se desea analizar el desempeño del sistema de espera y responder las siguientes preguntas:

1. Tiempo de espera promedio en la fila antes de ser atendido.

2. Tiempo total promedio en el sistema (incluyendo servicio).

3. Número promedio de clientes en el sistema en cualquier instante.

4. Probabilidad de que un cliente tenga que esperar (es decir, todos los servidores están ocupados).

5. Nivel de utilización del sistema (ratio entre la tasa de llegada y la capacidad total de servicio).

```python
from math import exp
from math import factorial
```

```python
#UtilizacionPromedio
def p(tasaLlegada, tasaServicio, servidores):
    return (float(tasaLlegada/(tasaServicio*servidores)))
```

```python
#p0
def pZero(tasaLlegada, tasaServicio, servidores):
    calc_parte1 = 0
    for x in range(servidores):
        calc_parte1 += ((tasaLlegada/tasaServicio)**x)/factorial(x)
    seccion1 = ((tasaLlegada/tasaServicio)**servidores/factorial(servidores))
    seccion2 = 1/(1-p(tasaLlegada, tasaServicio, servidores))
    result_seccion2 = seccion1 * seccion2
    return 1/ (calc_parte1 + result_seccion2)
```

```python
#NUMERO ESPERADO DE CLIENTES EN COLA (LQ)
def lq(tasaServicio, tasaLlegada, servidores):
    num = pZero(tasaLlegada, tasaServicio, servidores) * ((tasaLlegada/tasaServicio)**servidores) * p(tasaLlegada, tasaServicio, servidores)
    den = factorial(servidores) * ((1-p(tasaLlegada, tasaServicio, servidores))**2)

    return num/den
```

```python
def l(tasaServicio, tasaLlegada, servidores):
    return lq(tasaServicio, tasaLlegada, servidores) + (tasaLlegada/tasaServicio)
```

```python
#TIEMPO PROMEDIO EN COLA
def wq(tasaServicio, tasaLlegada, servidores):
    return lq(tasaServicio, tasaLlegada, servidores) / tasaLlegada  
```

```python
def w(tasaServicio, tasaLlegada, servidores):
    return wq(tasaServicio, tasaLlegada, servidores) + (1/tasaServicio)
```

```python
def Pw(tasaServicio, tasaLlegada, servidores):
    mult1 = 1/factorial(servidores)
    mult2 = (tasaLlegada/tasaServicio)**servidores
    mutl3 = (servidores*tasaServicio)/(servidores*tasaServicio - tasaLlegada)
    mult4 = pZero(tasaLlegada, tasaServicio, servidores)
    return mult1 * mult2 * mutl3 * mult4
```

```python
tasaLlegada = int(input("Ingrese la tasa de llegadas por hora: "))
tasaServicio = int(input("Ingrese el ritmo de servicio por hora: "))
servidores = int(input("Ingrese el numero de servidores: "))
```

```python
print("Tiempo de espera promedio en la fila antes de ser atendido.", wq(tasaServicio, tasaLlegada, servidores))
```

```python
print("Tiempo total promedio en el sistema (incluyendo servicio).", w(tasaServicio, tasaLlegada, servidores))
```

```python
print("Número promedio de clientes en el sistema en cualquier instante.", l(tasaServicio, tasaLlegada, servidores))
```

```python
print("Probabilidad de que un cliente tenga que esperar (es decir, todos los servidores están ocupados).", Pw(tasaServicio, tasaLlegada, servidores))
```

```python
print("5. Nivel de utilización del sistema (ratio entre la tasa de llegada y la capacidad total de servicio).", p(tasaLlegada, tasaServicio, servidores))
```
