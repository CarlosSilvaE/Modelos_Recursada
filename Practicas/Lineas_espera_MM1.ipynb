La gerente de una tienda de abarrotes, esta interesada en brindar un buen servicio a las personas mayores que compran en su tienda. Actualmente, la tienda tiene una caja registradora reservada para los clientes de la tercera edad. Esas personas llegan a la caja a un ritmo promedio de 30 clientes por hora, de acuerdo con una distribución de Poisson, y son atendidos a una tasa promedio de 35 clientes por hora, con tiempos de servicios exponenciales.

```python
# A) CALCULE LA PROBABILIDAD DE QUE HAYA CERO CLIENTES EN EL SISTEMA?
# B) DETERMINE LA UTILIZACION PROMEDIO DEL EMPLEADO EN LA CAJA REGISTRADORA
# C) DETERMINE EL NUMERO PROMEDIO DE CLIENTES EN EL SISTEMA
# D) DETERMINE EL NUMERO PROMEDIO DE CLIENTES FORMADOS EN FILA
# E) CALCULE TIEMPO PROMEDIO QUE LOS CLIENTES PASAN EN EL SISTEMA
# F) CALCULE TIEMPO PROMEDIO DE ESPERA EN FILA
```

```python
#UTILIZACION PROMEDIO P
def P(tasaLlegada, tasaServicio):
    return (float(tasaLlegada/tasaServicio)*100)
```

```python
#NUMERO ESPERADO DE CLIENTES EN COLA (LQ)
def lq(tasaServicio, tasaLlegada):
    return (float((tasaLlegada * tasaLlegada)/(tasaServicio*(tasaServicio-tasaLlegada))))
```

```python
#NUMERO ESPERADO DE CLIENTES RECIBIENDO EL SERVICIO(LS)
def ls(tasaServicio, tasaLlegada):
    return tasaLlegada/tasaServicio
```

```python
#NUMERO ESPERADO DE CLIENTES EN EL SISTEMA DE COLA(LW)
def l(tasaServicio, tasaLlegada):
    return tasaLlegada/(tasaServicio - tasaLlegada)
```

```python
#PROBABILIDAD DE QUE EL SISTEMA ESTE VACIO
def p0(tasaServicio, tasaLlegada):
    return 1-(tasaLlegada/tasaServicio)
```

```python
#TIEMPO PROMEDIO DE PERMANENCIA EN EL SISTEMA
def w(tasaServicio, tasaLlegada):
    return 1/(tasaServicio/tasaLlegada)
```

```python
#TIEMPO PROMEDIO EN COLA
def wq(tasaServicio, tasaLlegada):
    return lq(tasaServicio, tasaLlegada)/tasaServicio
```
