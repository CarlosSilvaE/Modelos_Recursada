```python
#Importacion para el uso de librerias
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
#Declaracion de datos a utilizar en vectores y matrices
nombres = ['Ana', 'Luis', 'Carlos', 'Marta']
calificaciones = [[5, 5.5, 10], [9, 8.7, 10], [3, 5, 7], [10, 10, 10]]
```

```python
#Crear dataframe
df = pd.DataFrame({'nombres': nombres, 'calificaciones': calificaciones})
df
```

```python
#Calculos
medias = []

for x in df['calificaciones']:
    medias.append(sum(x)/len(x))

df['media'] = medias
cnt = 0
```

```python
df
```

```python
plt.bar(df['nombres'], df['media'], color="blue")
plt.xlabel('Estudiantes')
plt.ylabel('Promedios')
plt.title('Promedio de calificaciones por estudiante')

for i in df['media']:
    plt.Text(cnt, i + 0.1, str(round(i,2)), ha="center")
    cnt+=1
plt.show()
```
