LA CERVECERIA MAS IMPORTANTE DEL MUNDO (GUINESS) HA CONTRATADO UN ANALISTA DE INVESTIGACION DE OPERACIONES PARA ANALIZAR SU POSICION EN EL MERCADO. ESTAN PREOCUPADOS EN ESPECIAL POR SU MAYOR COMPETIDOR (HEINEKEN). EL ANALISTA PIENSA QUE EL CAMBIO DE MARCA SE PUEDE MODELAR CON UNA CADENA DE MARKOV INCLUYENDO 3 ESTADOS, LOS ESTADOS G Y H REPRESENTAN A LOS CLIENTES QUE BEBEN CERVEZA PRODUCIDA POR LAS MENCIONADAS CERVECERIAS Y EN EL ESTADO I REPRESENTA A TODAS LAS DEMAS MARCAS. LOS DATOS SE TOMAN CADA MES Y EL ANALISTA A CONSTRUIDO LA SIGUIENTE MATRIZ DE TRANSICION DE DATOS HISTORICOS.

```python
""""
   G    H     I
G[0.7, 0.2, 0.1]
H[0.2, 0.75, 0.05]
I[0.1, 0.1, 0.8]
"""
#¿CUALES SON LOS PORCENTAJES DE MERCADO EN EL ESTADO ESTABLE PARA LAS DOS CERVECERIAS MAS GRANDES?

import numpy as np
m1 = np.matrix([[0.7, 0.2, 0.1],[0.2, 0.75, 0.05], [0.1, 0.1, 0.8]])
m1
```

```python
m_aux = np.matrix([[], [], []])
m_aux_anterior = np.matrix([[], [], []])
```

```python
#DETECTAR MATRIZ ESTABLE
for i in range(1000):
    if i == 0:
        m_aux = np.dot(m1, m1)
        print(m_aux)
        print("1-------------", i + 1)
    else:
        y = i + 1
        m_aux_anterior = m_aux
        m_aux = np.dot(m_aux, m1)
        print(m_aux)
        if (np.array_equal(m_aux, m_aux_anterior)):
            break
        print(y, "-------", i + 1)
print()
print("La matriz estable es:\n", np.round(m_aux, 4))
print("-------------------")
print(np.round(m_aux, 3))
```

```python
print("El porcentaje para guinness en el mercado es:", round(m_aux[2,0] * 100, 2), "%")

print("El porcentaje para heineken en el mercado es:", round(m_aux[2,1] * 100, 2), "%")

print("El porcentaje para las demas en el mercado es:", round(m_aux[2,2] * 100, 2), "%")
```

```python
#DETERMINAR VALOR DE LA POBLACION
poblacion_inicial = np.matrix([6000,3000,1000])
poblacion_inicial
```

```python
print("Poblacion inicial----V0")
for z in range(50):
    if z == 0:
        poblacion_inicial = np.dot(poblacion_inicial, m1)
        poblacion_inicial = np.round(poblacion_inicial, 2)
        print("Poblacion inicial")
        print("-------------", z+1)
    else:
        poblacion_anterior = poblacion_inicial
        poblacion_inicial = np.dot(poblacion_inicial, m1)
        poblacion_inicial = np.round(poblacion_inicial, 2)
        if (np.array_equal(poblacion_inicial, poblacion_anterior)):
            print("Poblacion inicial y anterior iguales")
            print("*** numero de iteraciones", str(z), "***")
            poblacion_inicial = np.round(poblacion_inicial, 0)
            poblacion_inicial = poblacion_inicial.astype(int)
            print("Poblacion final:\n", poblacion_inicial)
            print("N ---------", z+1)
            break
print(poblacion_inicial)
print("La poblacion se estabiliza en la iteracion", z+1)
```
