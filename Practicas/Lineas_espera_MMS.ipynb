UNA DULCERIA DE CINE TIENE TRES EMPLEADOS QUE ATIENDEN A LOS CLIENTES POR RIGUROSO ORDEN DE LLEGADA. EL TIEMPO DE SERVICIO POR CLIENTE SE DISTRIBUYE EXPONENCIALMENTE , CON UN PROMEDIO DE 2 MINUTOS POR CLIENTE. LOS CLIENTES DE LA DULCERIA ESPERAN EN UN AMPLIO ESPACIO FORMANDO UNA SOLA FILA Y SUS LLEGADAS TIENEN UNA DISTRIBUCION DE POISSON CON UN PROMEDIO DE 81 CLIENTES POR HORA.
SI EL TIEMPO PROMEDIO QUE PASAN EN EL AREA DE DULCERIA REBASA LOS 10 MINUTOS LOS CLIENTES SE SIENTEN INSATISFECHOS.

```python
from math import exp
from math import factorial
```

```python
#UtilizacionPromedio
def p(tasaLlegada, tasaServicio, servidores):
    return (float(tasaLlegada/(tasaServicio*servidores))*100)
```

```python
#p0
def pZero(tasaLlegada, tasaServicio, servidores):
    calc_parte1 = 0
    for x in range(servidores):
        calc_parte1 += ((tasaLlegada/tasaServicio)**x)/factorial(x)
    seccion1 = ((tasaLlegada/tasaServicio)**servidores/factorial(servidores))
    seccion2 = 1/(1-p(tasaLlegada, tasaServicio, servidores))
    result_seccion2 = seccion1 * seccion2
    return 1/ (calc_parte1 + result_seccion2)
```
