```python
import numpy as np
```

```python
m1 = np.matrix([[0.80, 0.15, 0.05],[0.3, 0.6, 0.1], [0.4, 0.3, 0.3]])
m1
```

```python
#FUNCIONANDO, CUAL ES LA PROBABILIDAD DE QUE AL DIA SIGUIENTE ESTE EN MANTENIMIENTO
#F, M. A
v3 = np.matrix([[1, 0, 0]])
v3_result = np.dot(v3, m1)

print("Cual es la probabilidad de que si hoy esta funcionando, mañana pase a mantenimiento", v3_result[0,1] * 100)
```

```python
# Probabilidad a dos dias de estar en funcionamiento a averiado

p2 = np.dot(m1, m1)

v4 = np.matrix([[1, 0, 0]])
v4_result = np.dot(v4, p2)

print("Cual es la probabilidad de que si hoy esta funcionando, mañana pase a averiada", round(v4_result[0,2] * 100, 0))
```

```python
#DETECTAR MATRIZ ESTABLE
for i in range(1000):
    if i == 0:
        m_aux = np.dot(m1, m1)
        print(m_aux)
        print("1-------------", i + 1)
    else:
        y = i + 1
        m_aux_anterior = m_aux
        m_aux = np.dot(m_aux, m1)
        print(m_aux)
        if (np.array_equal(m_aux, m_aux_anterior)):
            break
        print(y, "-------", i + 1)
print()
print("La matriz estable es:\n", np.round(m_aux, 4))
print("-------------------")
print(np.round(m_aux, 4))
```

```python
vector_estable = np.round(m_aux[0],4)

#CONVERTIRLO A PORCENTAJE
porcentaje = vector_estable * 100
print("Distribucion estable estacionaria en %")
print(f"Funcionando:{porcentaje[0,0]:.2f}%")
print(f"Mantenimiento:{porcentaje[0,1]:.2f}%")
print(f"Averiada:{porcentaje[0,2]:.2f}%")
```
