# CADENA DE MARKOV

Se asume que tenemos un puesto de restaurante donde solo servimos tres tipos de comida que son <br> 
hamburguesa, pizza y hot dog, pero tienen un patron especifico, solo sirven uno de estas tres comidas por dia y <br>
depende de lo que hayan servido el dia anterior, en terminos generales se busca la manera de predecir lo que <br>
serviran mañana si sabemos lo que sirvieron hoy y las ganancias estimadas a obtener. <br>
Por ejemplo, hay una probabilidad de 60% de que mañana sea un dia de pizza dado que hoy es una <br> 
hamburguesa, hay una probabilidad de 20% de que mañana vuelvan a servir hamburguesas y la probabilidad <br>
restante es para transitar a hot dog. <br>
En el caso de la pizza es practicamente nulo que se repita, pero puede pasar al menu del dia en un 30% para <br>
hamburguesa y un 70% para hot dog. <br>
El hot dog es de los productos mas vendidos y se puede repetir en un 50%, por reglas del negocio no puede <br>
pasar a ventas de pizza, pero si puede transitar en el porcentaje restante al menu de hamburguesas.

Se desea determinar <br>
- El restaurante sirvio pizza el primer dia y hamburguesa el segundo, nuevamente pizza el tercer dia, ¿Cual <br> 
es la probabilidad de que sirvan hot dogs el cuarto dia?

la ganancia estimada por cada tipo de alimento es de 40 pesos, 50 pesos y 25 pesos respectivamente, <br>
¿Cual sera la utilidad mensual(30 dias) esperada de ganancias para el negocio acorde a la cadena de markov?

# LINEAS DE ESPERA (COLAS)

```python
from math import factorial
```

```python
#UtilizacionPromedio
def p(tasaLlegada, tasaServicio, servidores):
    return (float(tasaLlegada/(tasaServicio*servidores)))
```

```python
#p0
def pZero(tasaLlegada, tasaServicio, servidores):
    calc_parte1 = 0
    for x in range(servidores):
        calc_parte1 += ((tasaLlegada/tasaServicio)**x)/factorial(x)
    seccion1 = ((tasaLlegada/tasaServicio)**servidores/factorial(servidores))
    seccion2 = 1/(1-p(tasaLlegada, tasaServicio, servidores))
    result_seccion2 = seccion1 * seccion2
    return 1/ (calc_parte1 + result_seccion2)
```

```python
#CANTIDAD PROMEDIO DE PACIENTES EN COLA (LQ)
def lq(tasaServicio, tasaLlegada, servidores):
    num = pZero(tasaLlegada, tasaServicio, servidores) * ((tasaLlegada/tasaServicio)**servidores) * p(tasaLlegada, tasaServicio, servidores)
    den = factorial(servidores) * ((1-p(tasaLlegada, tasaServicio, servidores))**2)

    return num/den
```

```python
#CANTIDAD PROMEDIO DE PACIENTES EN EL SISTEMA
def l(tasaServicio, tasaLlegada, servidores):
    return lq(tasaServicio, tasaLlegada, servidores) + (tasaLlegada/tasaServicio)
```

```python
#TIEMPO PROMEDIO EN COLA
def wq(tasaServicio, tasaLlegada, servidores):
    return lq(tasaServicio, tasaLlegada, servidores) / tasaLlegada  
```

```python
def w(tasaServicio, tasaLlegada, servidores):
    return wq(tasaServicio, tasaLlegada, servidores) + (1/tasaServicio)
```

```python
tasaLlegada = 12
tasaServicio = 9
servidores = 3
```

```python
#¿Cantidad promedio de pacientes en cola?
print("Cantidad promedio de pacientes en cola: ", lq(tasaServicio, tasaLlegada, servidores))

#¿Cantidad promedio de pacientes en el sistema?
print("Cantidad promedio de pacientes en el sistema: ", l(tasaServicio, tasaLlegada, servidores))

#¿Tiempo promedio en cola?
print("Tiempo promedio en cola: ", wq(tasaServicio, tasaLlegada, servidores))

#¿Tiempo promedio en el sistema?
print("Tiempo promedio de permanencia en el sistema: ", w(tasaServicio, tasaLlegada, servidores))
```
